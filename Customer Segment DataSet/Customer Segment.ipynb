## # Cleaing and Preparing Dataset.

### ---} Train Data.

```python
from sklearn.preprocessing import LabelEncoder


tr['Var_1'] = np.where(tr['Var_1'].isin(['Cat_1', 'Cat_2', 'Cat_3', 'Cat_5', 'Cat_7']), 'Others', tr['Var_1'])

tr_c = tr.dropna()
tr_cat = ["Gender","Ever_Married","Graduated","Profession","Spending_Score","Var_1"]


le = LabelEncoder()

for col in tr_cat:
    tr_c[col] = le.fit_transform(tr_c[col])

tr_new = tr_c.sample(frac = 1, random_state=42)
```

```python
tr_clean = tr_new
```

```python
tr_clean.head()
```

### ---} Test Data.

```python
from sklearn.preprocessing import LabelEncoder


te['Var_1'] = np.where(te['Var_1'].isin(['Cat_1', 'Cat_2', 'Cat_3', 'Cat_5', 'Cat_7']), 'Others', te['Var_1'])

te_c = te.dropna()
te_cat = ["Gender","Ever_Married","Graduated","Profession","Spending_Score","Var_1"]


le = LabelEncoder()

for col in te_cat:
    te_c[col] = le.fit_transform(te_c[col])

te_new = te_c.sample(frac = 1, random_state=42)
```

```python
te_clean = te_new
```

```python
te_clean.head()
```

# ---------------------------------------------------------------------------------------------------------------

## # Performing Train Test Split Over Dataset.

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Here Performing Train Test Split Over Training Dataset.

x = tr_clean.drop(["ID","Var_1","Segmentation"] , axis=1)
y = tr_clean["Var_1"]

y = to_categorical(y)

# Now Performing Train & Validation Split.

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.20, shuffle = True)
```

# ---------------------------------------------------------------------------------------------------------------

## # Establising Precision, Recall & F1 score function.

```python
# Creating Precision, Recall and F1 Function.

def precision_tr(y_true, y_pred):
    true_pos = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    p_pos = K.sum(K.round(K.clip(y_pred, 0, 1)))
    return true_pos / (p_pos + K.epsilon())

def recall_tr(y_true, y_pred):
    true_pos = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    r_pos = K.sum(K.round(K.clip(y_true, 0, 1)))
    return true_pos / (r_pos + K.epsilon())

def f1_val_tr(y_true, y_pred):
    p = precision_tr(y_true, y_pred)
    r = recall_tr(y_true, y_pred)
    return 2 * (p * r) / (p + r + K.epsilon())

```

# ---------------------------------------------------------------------------------------------------------------

## # Designing a Neural Network Model.

```python
from keras import models
from keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
from keras import backend as K

m1 = models.Sequential()

# Add the first (input) layer (64 nodes) with input shape 12 element vector (1D).
m1.add(layers.Dense(64, activation='relu', input_shape=(8,)))

# Add the second (hidden) layer of 128 nodes.
m1.add(layers.Dense(128, activation='relu'))

# Add the Third (hidden) layer of 128 nodes.
m1.add(layers.Dense(128, activation='relu'))

# Add the Forth (hidden) layer of 3 nodes.
m1.add(layers.Dense(3, activation='sigmoid'))

```

```python
m1.summary()
```

```python
# compiling the model 1.
m1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', precision_tr, recall_tr, f1_val_tr])

# Early stopping will help to stop the model at 150 instead of given epochs

es = EarlyStopping(patience=150)

# Fitting the model
history = m1.fit(x_train, y_train, epochs=400, batch_size=50,validation_data=(x_val, y_val),verbose=1, callbacks = [es])
```

### ---} Training Data.

```python
tr_acc = m1.evaluate(x_train, y_train, verbose = 0)
tr_f1 = m1.evaluate(x_train, y_train, verbose = 0)
tr_r = m1.evaluate(x_train, y_train, verbose = 0)
tr_p = m1.evaluate(x_train, y_train, verbose = 0)
tr_loss = m1.evaluate(x_train, y_train, verbose = 0)

print(" ---> Accuracy : {} \n ---> F1 Score : {} \n ---> Recall : {} \n ---> Precission : {} \n ---> Loss : {} ".format(tr_acc, tr_f1, tr_r, tr_p, tr_loss))
```

### ---} Validation Data.

```python
val_acc = m1.evaluate(x_val, y_val, verbose=0)
val_f1 = m1.evaluate(x_val, y_val, verbose=0)
val_r = m1.evaluate(x_val, y_val, verbose=0)
val_p = m1.evaluate(x_val, y_val, verbose=0)
val_loss = m1.evaluate(x_val, y_val, verbose=0)

print(" ---> Accuracy : {} \n ---> F1 Score : {} \n ---> Recall : {} \n ---> Precission : {} \n ---> Loss : {} ".format(val_acc, val_f1, val_r, val_p, val_loss))
```
